```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
import statsmodels.api as sm
```

```python
mat_df = pd.read_csv('new_student-mat.csv')
mat_df.head(10)
```

```python
mat_df.isnull().sum()
```

```python
por_df = pd.read_csv('new_student-por.csv')
por_df.head(10)
```

```python
por_df.isnull().sum()
```

```python
df = pd.concat([mat_df,por_df])
df.head(10)
```

```python
#add grade to the final grade into categorical variable(good, fair and poor)
df['final_grade'] = 'na'
df.loc[(df.G3 >= 15)& (df.G3 <=20), 'final_grade'] = 'good'
df.loc[(df.G3 >= 10)& (df.G3 <=14), 'final_grade'] = 'fair'
df.loc[(df.G3 >= 0)& (df.G3 <=9), 'final_grade'] = 'poor'
df
```

<h2 style = 'text-align: center'>Data Analysis</h2>

```python
#Frequency of the final_grade
df['final_grade'].value_counts().plot(kind = 'bar')
plt.title('Final Grade - Number of Students',fontsize=20)
plt.xlabel('Final Grade', fontsize=16)
plt.ylabel('Number of Student', fontsize=16)
```

```python
#Correlation heatmap
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
num_df = df[numerical_cols]

corr_matrix = num_df.corr()

sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt='.2f', annot_kws={'fontsize': 8})
```

```python
#relationship vs no relationship students
index = ['poor','fair','good']
romance_tab = pd.crosstab(index=df.final_grade, columns=df.romantic)
romance = romance_tab.reindex(index)
```

```python
plt.figure()
romance.plot.bar(colormap="PiYG_r", fontsize=16, figsize=(8,8))
plt.title('Final Grade By Romantic Status', fontsize=20)
plt.ylabel('Number of Student Counts ', fontsize=16)
plt.xlabel('Final Grade', fontsize=16)
plt.show()
```

```python
#Test hyphothesis
import statsmodels.api as sm
romance_table = sm.stats.Table(romance_tab)
romance_rslt = romance_table.test_nominal_association()
romance_rslt.pvalue
print('p-value:',romance_rslt.pvalue,'\nHyphothesis Testing confirmed, romantic status has a significant correlation with final grade.')
```

```python
#Final Grade By Alcohol Consumption
perc = (lambda col: col/col.sum())
alc_tab1 = pd.crosstab(index=df.final_grade, columns=df.Walc)
alc_tab = np.log(alc_tab1)
alc_perc = alc_tab.apply(perc).reindex(index)
```

```python
alc_perc.plot.bar(colormap="Reds", figsize=(10,8), fontsize=16)
plt.title('Final Grade By Weekend Alcohol Consumption', fontsize=20)
plt.ylabel('Percentage of Logarithm Student Counts', fontsize=16)
plt.xlabel('Final Grade', fontsize=16)
```

```python
#Test hyphothesis
alc_table = sm.stats.Table(alc_tab1)
alc_rslt = alc_table.test_nominal_association()
print('p-value:',alc_rslt.pvalue,'\nHyphothesis Testing confirmed, weekend alcohol consumption has a significant correlation with final grade.')
```

```python
#Final Grade By Parents Education Level
X_edu = df[['Medu','Fedu']]
y_edu = df.G3
edu = sm.OLS(y_edu, X_edu)
results_edu = edu.fit()
results_edu.summary()
```

OLS tells that parents' education level has a positive correlation with students' final score. Comparatively, mother's education level has bigger influence than father's education level

<h2 style = 'text-align: center'>Model</h2>

***Multi Linear Regression***

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn import preprocessing
data = df

le = preprocessing.LabelEncoder()
data.school = le.fit_transform(data.school)
data.sex = le.fit_transform(data.sex)

X1 = data[['school','sex','age','Medu','Fedu','studytime','freetime','goout','Walc','health','absences','G1','G2']]
y1 = data['G3']

# Add a constant to the independent variables for Model
X1 = sm.add_constant(X1)

# Calculate the VIF values for Model
vif1 = pd.DataFrame()
vif1["VIF Factor"] = [variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])]
vif1["features"] = X1.columns

# Fit Model using OLS regression
model1 = sm.OLS(y1, X1).fit()

# Print the summary of Model and the calculated VIF values
print("Model 1(Multi Linear Regression):")
print(model1.summary())
print("VIF Values:")
print(vif1)
```

***Logistic Regression***

```python
df_model = df.copy()
df_model = df_model.drop(['G3'], axis=1)
```

```python
#Encode final_grade
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df_model.final_grade = le.fit_transform(df_model.final_grade)
```

```python
#train_test_split dataset
from sklearn.model_selection import train_test_split
X = df_model.drop('final_grade',axis=1) 
y = df_model.final_grade
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
# get dummy varibles 
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(multi_class='multinomial', solver='newton-cg',fit_intercept=True)
```

```python
# find optimal # of features to use in the model
from sklearn.feature_selection import SelectKBest, chi2

ks=[]
for i in range(1,57):
    sk = SelectKBest(chi2, k=i)
    x_new = sk.fit_transform(X_train,y_train)
    x_new_test=sk.fit_transform(X_test,y_test)
    l = lr.fit(x_new, y_train)
    ll = l.score(x_new_test, y_test)
    ks.append(ll)  
    
ks = pd.Series(ks)
ks = ks.reindex(list(range(1,57)))
```

```python
sk = SelectKBest(chi2, k=8)
x_new = sk.fit_transform(X_train,y_train)
x_new_test=sk.fit_transform(X_test,y_test)
lr = lr.fit(x_new, y_train)
print("Logistic Regression Model Score" , ":" , lr.score(x_new, y_train))
print('Model parameters:', lr.get_params())
print('Model coefficients:', lr.coef_)
print('Model intercept:', lr.intercept_)
```
